```python
categorical_cols = df.select_dtypes(include='object').columns

```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
numlist=['Age','Income','LoanAmount',	'CreditScore','MonthsEmployed','NumCreditLines','InterestRate','LoanTerm','DTIRatio']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df[numlist]=scaler.fit_transform(df[numlist])
```

```python
df = df.drop(columns=["LoanID"])
```

```python
df.head()
```

```python
x=df.drop(columns=['Default'])
y=df['Default']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight="balanced", max_iter=2000)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=400,
    class_weight="balanced",
    max_depth=None,
    random_state=42,
    n_jobs=-1
)

rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)
```

```python
confusion_matrix(y_test, y_pred_rf)
```

```python
# Step 1: Calculate imbalance ratio
neg = y_train.value_counts()[0]   # SAFE
pos = y_train.value_counts()[1]   # RISKY

scale = neg / pos                 # safe / risky ratio
scale
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.1,
    reg_lambda=1,
    scale_pos_weight=scale,   # <------- THIS IS WHERE scale is USED
    random_state=42,
    n_jobs=-1,
    tree_method="hist"
)
```

```python
xgb.fit(x_train, y_train)
```

```python
y_pred_xgb = xgb.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_xgb))
```

```python
confusion_matrix(y_test, y_pred_xgb)
```

**Get** **probabilities**:

```python
y_proba_xgb = xgb.predict_proba(x_test)[:, 1]
```

**Choosing ANY threshold**:

```python
threshold = 0.30
```

**Convert probability to prediction using your threshold**:

```python
y_pred_thr = (y_proba_xgb >= threshold).astype(int)
```

**Check recall + confusion matrix**:

```python
from sklearn.metrics import recall_score, confusion_matrix

print("Threshold:", threshold)
print("Recall:", recall_score(y_test, y_pred_thr))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_thr))
```

```python
final_threshold = 0.30

y_proba_final = xgb.predict_proba(x_test)[:, 1]
y_pred_final = (y_proba_final >= final_threshold).astype(int)
```
